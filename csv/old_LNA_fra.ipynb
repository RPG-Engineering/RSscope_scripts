```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy

# Fetch data
columns = ["freq", "Vppin", "Vppout"]
raw=pd.read_csv("https://raw.githubusercontent.com/RPG-Engineering/RSscope_scripts/main/csv/20230201-135537_MXO44vsOldPreamp.csv", usecols=columns)
plt.rcParams["figure.figsize"] = [14, 7]
fig, ax = plt.subplots()
plt.xscale("log")
ax.plot(raw["freq"], raw["Vppout"])
ax.set(xlabel='frequency (Hz)', ylabel='Voltage peak-peak (V)',
       title='Old 0.1 - 10 Hz preampt frequency response with 1 Vpp input')
ax.grid()
plt.show()
```
